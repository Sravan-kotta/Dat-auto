```python
def null_value_treat_impute(file_path):
    data = pd.read_csv(file_path)
    for column in data:
         if data.dtypes[column] == "datetime64":
            print ("time series detected in dataset:",column,'''suggested methods: [5]forward fill
                                                                                   [6]backward fill''')
    print ("select method for individual columns, please do note suggested methods")
    for column in data: 
        if data.dtypes[column] == "object":
            StrChk = int(input("string type detected for",column,''' skip or apply label encoding?(for catagorical) : [1]skip
                                                                                                                 [2]label encoding''' )) 
            if StrChk == 1:
                continue
            if StrChk == 2:
                l1 = LabelEncoder()
                l1.fit(data[column])
                data[column] = l1.transform(data[column])   
        elif data.dtypes[column] == "float64" or data.dtypes[column] == "int64":
            print ("numerical data detected for column:",column,'''suggested methods: [2]mean 
                                                                                      [3]median (skewed data)
                                                                                      [4]mode (catagorical data)''')
        x = int(input('''choose method from following:
                        1 for constant impute
                        2 for mean impute
                        3 for median impute
                        4 for mode impute
                        5 for forward fill impute
                        6 for backward fill impute'''))
        if x == 1:
            c = float(input("enter constant:"))
            data[column] = data[column].fillna(c, inplace=False)
        elif x == 2:
            imputer = SimpleImputer(fill_value=np.nan, startegy='mean')
            data[column] = imputer.fit_transform(data[column]) 
        elif x == 3:
            imputer = SimpleImputer(fill_value=np.nan, startegy='median')
            data[column] = imputer.fit_transform(data[column])
        elif x == 4:
            imputer = SimpleImputer(fill_value=np.nan, startegy='mode')
            data[column] = imputer.fit_transform(data[column])  
        elif x == 5:
            data[column] = data[column].ffill(axis = 0)
            if data[column].isnull().iloc[0] == True:
                w = int(input('''first row is empty, fill with:  1 for constant impute
                                                          2 for mean impute
                                                          3 for median impute
                                                          4 for mode impute''')) 
                if x == 1:
                    d = float(input("enter constant:"))
                    data[column] = data[column].fillna(d) 
                elif x == 2:
                    data[column] = data[column].fillna(data[column].mean())  
                elif x == 3:
                    data[column] = data[column].fillna(data[column].median())
                elif x == 4:
                    data[column] = data[column].fillna(data[column].mode()) 
            x = data[column].isnull().mean()*100
            print ("null value in",column,":",x,"%")
        elif x == 6:
            data[column] = data[column].bfill(axis = 0)
            if data[column].isnull().iloc[-1] == False:
                w = int(input('''last row is empty, fill with:  1 for constant impute
                                                          2 for mean impute
                                                          3 for median impute
                                                          4 for mode impute''')) 
          
                if x == 1:
                    e = float(input("enter constant:"))
                    data[column] = data[column].fillna(e) 
                elif x == 2:
                    data[column] = data[column].fillna(data[column].mean())  
                elif x == 3:
                    data[column] = data[column].fillna(data[column].median())
                elif x == 4:
                    data[column] = data[column].fillna(data[column].mode())
            x = data[column].isnull().mean()*100
            print ("null value in",column,":",x,"%")
                        
```
